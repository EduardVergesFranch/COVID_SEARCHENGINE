# Hydrate Tweets

This script will load further information about our collected tweets such as RT and FAV count. This is important because we stream real time tweets without interactions at first.

## Important notes

#### Download required files stored in Google Drive or execute cleaning process
This script requires the file `tweets.json` that can be downloaded from the Google Drive Folder (`https://drive.google.com/drive/u/1/folders/16I4_ZCre59ufD9lDZbFK9cn1mALRmPjB`). The file must be stored in the `~/data` folder as specified in the README.

Otherwise, you can execute the `Crawler.ipynb` script.

```python
import twarc
import pandas as pd
from chronometer import Chronometer
import json
```

```python
input_path = '../../data/'

with open(os.path.join(input_path, 'tweets.json'), "rb") as f:
    data = f.readlines()
    data = [json.loads(str_) for str_ in data]
```

```python
df_tweets = pd.DataFrame.from_records(data)
```

```python
T = twarc.Twarc()

# retrieve original tweets from their ids
def hydrate(ids):
    last_t = 0
    with Chronometer() as t:
        count = 0
        hydrated_tweets = []

        for tweet in T.hydrate(iter(ids)):
            assert tweet['id_str']
            count += 1
            hydrated_tweets.append(tweet)
            if(int(float(t)) % 10 == 0 and int(float(t)) != last_t):
                    print("Hydrated tweets:", len(hydrated_tweets))
                    last_t = int(float(t))

        with open(input_path + "hydrated_tweets.json", 'a') as json_file:
            json.dump(hydrated_tweets, json_file)
    
    return count

hydrate(list(df_tweets['id_str'].values))
```
